## 条件付き期待値の計算

簡単な線形モデル
\begin{eqnarray}
y = x + v\\
x～N(\overline{x},\sigma^2_a)\\
v～N(0,\sigma^2_v)
\end{eqnarray}

### 正規分布の加法性

\begin{eqnarray}
a ～ N(\mu_a, \sigma_a^2)\\
b ～ N(\mu_b, \sigma_b^2)\\
であるとき\\
a + b も正規分布に従い、\\
a + b ～ N(\mu_a+\mu_b, \sigma_a^2+\sigma_b^2)
となる。
\end{eqnarray}

### [数値例] x

```python
import numpy as np
np.random.seed(0)
N = 1000
```

```python
#
x_ave = 10
x_dev = 2
x_var = x_dev**2
x = np.random.normal(x_ave, x_dev, N)

#
import matplotlib.pyplot as plt
plt.figure(figsize=(14, 3))
plt.subplot(1, 2, 1)
plt.plot(x, '.')
plt.subplot(1, 2, 2)
plt.hist(x, bins=20)
plt.show()
```

### [数値例] y

```python
v_ave = 0
v_dev = 2
v_var = v_dev**2
v = np.random.normal(v_ave, v_dev, N)
y = x + v

print(np.mean(y))
print(np.cov(y))
print(np.sqrt(np.cov(y)))
```

### 事後確率密度と条件付き期待値

#### ベイズの定理

\begin{eqnarray}
p(x_i \mid y_i) &=&
\frac
    {集合「x=xi」における、y=yiとなる事象}
    {集合「全x」における、y=yiとなる事象全て}
\\&=& 
\frac
    {p(y_i \mid x_i)p(x_i)}
    {\int p(y_i \mid x)p(x)dx}
\end{eqnarray}

事後確率密度関数は
\begin{eqnarray}
p(x \mid y) =
\frac
    {1}
    {\sqrt{2 \pi \tau^2}}
\exp(
    -
    \frac{1}{2 \tau^2}
    (x - \alpha \tau^2)^2
)
\end{eqnarray}

ここで
\begin{eqnarray}
\frac{1}{\tau^2} =
\frac{1}{\sigma_a^2} + 
\frac{1}{\sigma_v^2}
\\
\alpha =
\frac{\overline{x}}{\sigma_a^2} + 
\frac{y}{\sigma_v^2}
\end{eqnarray}

よって条件付き期待値は
\begin{eqnarray}
\alpha \tau ^2 = 
\frac{1}{\sigma_a^2+\sigma_v^2}
(\sigma_v^2 \overline{x} + \sigma_a^2 y)
\end{eqnarray}

#### [数値例] 事後確率分布

```python
tau2_inv = 1 / x_var + 1 / v_var
tau2 = 1 / tau2_inv

at2_denominator = x_var + v_var
at2_numerator1 = v_var * x_ave

xx = np.arange(0, 20, 0.2)

# y=10のとき
yi = 10
at2_numerator2 = x_var * yi
at2 = (at2_numerator1 + at2_numerator2) / at2_denominator
px_with_y = 1 / np.sqrt(2 * np.pi * tau2) * np.exp(-1 * (xx - at2)**2 / 2 * tau2 )

plt.plot(xx,px_with_y)
plt.title('p(x|y=10)')
plt.show()
```

```python
xx = np.arange(5, 15, 0.1)

plt.figure(figsize=(15,3))
#
yi = 10
at2_numerator2 = x_var * yi
at2 = (at2_numerator1 + at2_numerator2) / at2_denominator
px_with_y = 1 / np.sqrt(2 * np.pi * tau2) * np.exp(-1 * (xx - at2)**2 / 2 * tau2 )
plt.plot(xx,px_with_y,label='yi='+str(yi))

#
yi = 13
at2_numerator2 = x_var * yi
at2 = (at2_numerator1 + at2_numerator2) / at2_denominator
px_with_y = 1 / np.sqrt(2 * np.pi * tau2) * np.exp(-1 * (xx - at2)**2 / 2 * tau2 )
plt.plot(xx,px_with_y,label='yi='+str(yi))

#
yi = 16
at2_numerator2 = x_var * yi
at2 = (at2_numerator1 + at2_numerator2) / at2_denominator
px_with_y = 1 / np.sqrt(2 * np.pi * tau2) * np.exp(-1 * (xx - at2)**2 / 2 * tau2 )
plt.plot(xx,px_with_y,label='yi='+str(yi))

#
plt.legend()
plt.show()
```

#### [数値例] 条件付き期待値

```python
# sample at a point
yi = y[0]
denominator = x_var + v_var
numerator = v_var*x_ave + x_var*yi
at2 = numerator / denominator
print(at2)
```

```python
# loop
denominator = x_var + v_var
numerator1 = v_var * x_ave
yy = np.arange(0,20,0.5)
at2 = np.array([np.nan]*len(yy))
for i, yi in enumerate(yy):
    numerator2 = x_var * yi
    at2[i] = (numerator1 + numerator2) / denominator
```

```python
#
print('横軸：yの観測値, 縦軸：E[x│y]')
plt.plot(yy, at2, '.')
plt.xlabel('x')
plt.ylabel('E[x│y]')
plt.show()
```

## 条件付き期待値の概念

\begin{eqnarray}
y = x + v,　x～N(\overline{x},\sigma^2_a),　v～N(0,\sigma^2_v)
\end{eqnarray}
における
\begin{eqnarray}
&& p(y \mid x)　と　p(x \mid y)
\end{eqnarray}

### yの条件付き期待値

\begin{eqnarray}
p(y \mid x)
\end{eqnarray}
については、xが確定されたなら、
\begin{eqnarray}
y = x + v
\end{eqnarray}
より、期待値がxで、分散がvの分散と等しい確率分布となることが分かる。よって
\begin{eqnarray}
p(y \mid x) = 
\frac {1} {\sqrt{ 2 \pi \sigma_v^2 }}
\exp(-
    \frac
        { (y - x)^2 }
        {2 \sigma_v^2}
)
\end{eqnarray}


### xの条件付き期待値

では
\begin{eqnarray}
p(x \mid y)
\end{eqnarray}
について同じ論法が使えるか？つまり、
<blockquote>
<font color="#708090">
yが確定されたなら、
\begin{eqnarray}
x = y - v
\end{eqnarray}
より、期待値がyで、分散がvの分散と等しい確率分布と考えられる?
</font>
</blockquote>

結論から言えばNO。<br>
<hr>
1.<br>
　まず、「x：原因」「y：結果」という違いがある。<br>
  x　+　v　→　y　であり、<br>
  y　-　v　→　x　ではない。
<hr>
2.<br>
　x=x1が確定した後にyを求めるケースでは、<br>
原因：x1 + ランダムノイズ　→ 結果：y より<br>
「yの平均がx1であることは自明」である。<br>
　一方、y=y1が確定した後にxを求めるケースでは、<br>
y1が生成された過程は未知であり、<br>
y=y1を満たすxとvの組み合わせは様々考えられる。<br>
よって、ただちに「xの平均がy1であることは自明」と言うことはできない。<br>
<hr>

#### 2乱数の和と条件付き期待値

状態xとノイズvとか考えるとややこしい。<br>
平均ゼロの2つ乱数v,wと、その和yについて考える。<br>

\begin{eqnarray}
y = v + w \\
v～N(0,\sigma_v^2) \\
w～N(0,\sigma_w^2) \\
\end{eqnarray}

たとえば、観測値y=4が得られたとする。<br>
このときv, wの条件付き期待値はそれぞれいくらか？<br>

\begin{eqnarray}
4 &=& v + w において、下記の様々なパターンが考えられる\\
4 &=& 1 + 3 \\
4 &=& 0.2 + 3.8 \\
4 &=& 3.3 + 0.7 \\
4 &=& 6 - 2 \\
\end{eqnarray}
どういった組み合わせ（vとwのバランス）が一番確率が高いかという問題となる。

このとき、分散が大きいほど、大きな値の出る確率は高い。<br>
ここで「vの分散 ＜ wの分散」の場合を考えてみると、下図のようになる。
ここでは、v=w=2よりも、v=1,w=3の方が確率が高そうな様子を示している。


要は、平均と分散が既知の確率分布を持つ2乱数v,wがあるとき、2乱数の和v+w=yの値から、
それぞれの乱数の事後期待値を求めることが出来る。

分散の大小バランスが、条件付き期待値p(v│y)を決める上で重要となる。

## 条件付き期待値の計算

下記モデル（ランダムウォーク+ノイズ）について、条件付き確率を考えてみる。
\begin{eqnarray}
y_t = x_t + 観察ノイズw_t\\
x_{t+1} = x_t + 状態ノイズu_t \\
t = 1,2,...,n \\
状態初期値:x_0 \\ 
w_t～N(0,\sigma_w^2)\\
u_t～N(0,\sigma_u^2)\\
\end{eqnarray}

### E(yt│xt)

\begin{eqnarray}
この場合には、E(y_t \mid x_t)を求めるのは容易である。\\
上式の定義より、x_tが確定した後の条件つき確率分布p(y_t \mid x_t)は\\
平均x_t、分散\sigma_w^2となり、期待値はx_tである。\\
\end{eqnarray}

### E(xt│yt)

\begin{eqnarray}
一方、E(x_t \mid y_t)は即座には求められない。\\
これは、y_t = x_t + w_t　において、\\
x_tなどの確率分布が不明なためである。
\end{eqnarray}

\begin{eqnarray}
ここで乱数x_tについて、\\
x_{t+1} = x_t + u_t より\\
x_1 = x_0 + u_1\\
x_2 = x_1 + u_2 = x_0 + u_1 + u_2\\
したがって\\
x_t = x_0 + u_1 + u_2 +...+ u_t\\
であるから、初期値x_0、\sigma_u^2が与えられれば、x_tの確率分布p(x_t)は確定する。\\
さらに、\sigma_w^2が与えられれば、p(y_t)も確定し、\\
p(x_t \mid y_t)ならびにE(x_t \mid y_t)を求めることができる。
\end{eqnarray}

\begin{eqnarray}
未知パラメーター　x_0, \sigma_u^2, \sigma_w^2 \\
については、仮の値を設定して尤度を計算し、\\
最も尤度が高くなる値を採用する（最尤法）。
\end{eqnarray}

では、未知パラメーターが所与だとして、確率分布はどのように求められるか？
\begin{eqnarray}
正規分布の加法性より、\\
p(x_n)は平均x_0、分散n\sigma_{u}^2の正規分布となる。\\
また、p(y_n)は平均x_0,分散n\sigma_{u}^2+\sigma_{w}^2の正規分布となる。
\end{eqnarray}

\begin{eqnarray}
ここで、y_nの値が確定した後の事後確率分布p(x_n \mid y_n)は、\\
ベイズ定理より、\\
p(x_n \mid y_n) = 
\frac
    {p(y_n \mid x_n)p(x_n)}
    {p(y_n)}
=
\frac
    {p(y_n \mid x_n)p(x_n)}
    {\int{
        p(y_n \mid x_n)p(x_n)dx
        }
    }\\
※事後確率においてはy_nは定数であることに注意\\
ここで、式中のp(y_n \mid x_n)は、x_nがある定数として決定した後の事後確率分布であるから、\\
平均x_n、分散\sigma_w^2の確率分布における定数y_nとなる確率として一意に決まる。\\
※p(y_n)は既に定式化されているので、ここでは積分形で求める必要はない。
\end{eqnarray}

以上より、
\begin{equation}
p(x_n \mid y_n)ならびに\\
E(x_n \mid y_n)=E[p(x_n \mid y_n)*x_n]
\end{equation}
を計算することが出来る。

## 過去データを含む条件付き期待値 E(x_t │ y_0,y_1,...,y_t)

\begin{eqnarray}
では、過去の観察値y全て（y_1,y_2,...,y_n）を使える場合、つまり、\\
E(x_n \mid y_0,y_1,...,y_n)　はどうなるか？　\\
\end{eqnarray}

### p(x2|y1,y2)の例

\begin{eqnarray}
簡単な例としてp(x_2 \mid y_1,y_2)を考えてみよう。\\
x_2,y_1,y_2は下記のような順で決定される。\\
x_1 &=& x_0 + u_0\\
y_1 &=& x_1 + w_1\\
x_2 &=& x_1 + u_1\\
y_2 &=& x_2 + w_2
\end{eqnarray}
ここで同時確率分布を考えると
\begin{eqnarray}
p(y_1,x_2,y_2)  &=& p(y_1,y_2)p(x_2 \mid y_1,y_2) \\
                &=& p(y_1,x_2)p(y_2 \mid y_1,x_2) \\
となり、これは以下のように変形できる\\
p(x_2 \mid y_1,y_2) &=& p(y_2 \mid x_2,y_1)\frac{p(y_1,x_2)}{p(y_1,y_2)}\\
\end{eqnarray}
ここで
\begin{eqnarray}
p(y_1,x_2) = p(y_1)p(x_2 \mid y_1)\\
p(y_1,y_2) = p(y_1)p(y_2 \mid y_1)\\
\end{eqnarray}
であることを利用すると、
\begin{eqnarray}
p(x_2 \mid y_1,y_2) &=& p(y_2 \mid x_2,y_1)\frac{p(x_2 \mid y_1)}{p(y_2 \mid y_1)}\\
\end{eqnarray}

## 定理・公式

\begin{eqnarray}
ここで、y_nの値が確定した後の事後確率分布p(x_n \mid y_n)は、\\
ベイズ定理より、\\
p(x_n \mid y_n) = 
\frac
    {p(y_n \mid x_n)p(x_n)}
    {p(y_n)}
=
\frac
    {p(y_n \mid x_n)p(x_n)}
    {\int{
        p(y_n \mid x_n)p(x_n)dx
        }
    }\\
\end{eqnarray}

条件付き確率とベイズ定理
\begin{eqnarray}
aとbの同時確率 &=& 事象aが起こる確率×事象aが起こる集合のうち事象bが起こる確率\\ 
              &=& 事象bが起こる確率×事象bが起こる集合のうち事象aが起こる確率\\
p(a,b) &=& p(a)p(b \mid a)\\
       &=& p(b)p(a \mid b)\\
これを変形すると、\\
p(b \mid a) = \frac{p(b)p(a \mid b)}{p(a)}\\
\end{eqnarray}

同時確率
\begin{eqnarray}
p(a)とp(b)が互いに独立なとき、\\
p(a,b) = p(a)p(b)\\
独立な例：サイコロAの出目の確率分布と、サイコロBの出目の確率分布\\
独立ではない例：サイコロAの出目の確率分布と、サイコロAとBの出目の和の確率分布\\
※和はAの出目に影響されるので独立ではない。同様にランダムウォークのy_tとy_{t-1}も独立ではない。
\end{eqnarray}
